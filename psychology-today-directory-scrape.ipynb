```python
queens_pages=[]

# FLushing
for x in range (1,4):
    link = f"https://www.psychologytoday.com/us/therapists/ny/flushing?category=children-6-to-10&sid=625b2b98d7614&page={x}"
    queens_pages.append(link)

# Ridgewood
link = "https://www.psychologytoday.com/us/therapists/ny/ridgewood?category=children-6-to-10"
queens_pages.append(link)

#Astoria
for x in range (1,13):
    link = f"https://www.psychologytoday.com/us/therapists/ny/astoria?sid=625b2c435c82c&page={x}"
    queens_pages.append(link)
    
# Forest Hills
for x in range(1,3):
    link = f"https://www.psychologytoday.com/us/therapists/ny/forest-hills?category=children-6-to-10&sid=625b2c85306ca&page={x}"
    queens_pages.append(link)

#Jamaica
link = "https://www.psychologytoday.com/us/therapists/ny/jamaica?category=children-6-to-10"
queens_pages.append(link)

# Jackson Heights
link = "https://www.psychologytoday.com/us/therapists/ny/jackson-heights?category=children-6-to-10"
queens_pages.append(link)

# Sunnyside
for x in range(1,27):
    link = f"https://www.psychologytoday.com/us/therapists/ny/sunnyside?category=children-6-to-10&sid=625b2d1c08842&zipdist=3&page={x}"
    queens_pages.append(link)
```

```python
# pt_entries = []

for link in tqdm(queens_pages):
    raw_html = requests.get(link,headers=headers).content
    doc = BeautifulSoup(raw_html, "html.parser")
    rows = doc.find_all(class_="results-row")
    for row in rows:
        data = {}

        link = row.find_all('a')[1]
        profile_link = link['href']
        name = link.get_text()
        title = row.find_all(class_='profile-subtitle')[0].get_text()
        phone = row.find_all(class_='results-row-mob')[0].get_text()
        address = row.find_all(class_='profile-location')[0].get_text()
        statement = row.find_all(class_='statements')[0].get_text()

        data['url'] = profile_link
        data['name'] = name.strip()
        data['title'] = title.strip()
        data['phone'] = phone.strip()
        data['address'] = address
        data['statement'] = statement

        pt_entries.append(data)
        df = pd.DataFrame(pt_entries)
        
    # Sleep
    time.sleep(random.randint(1, 4))
```
